```python
# Your answer here (python code + explanation):
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

census=pd.read_csv("lab_assignment_05_census.csv")      #reading csv file into dataframe
```

```python
#acquiring relevant data from the dataframe
labels=["0-4","5-9","10-14","15-19","20-24","25-29","30-34","35-39","40-44","45-49","50-54","55-59","60-64","65-69","70-74","75-79","80-84","85-89","90+"]
man_19=census[census['Sex']==1][census['Year']==1900]['People']
women_19=census[census['Sex']==2][census['Year']==1900]['People']
man_20=census[census['Sex']==1][census['Year']==2000]['People']
women_20=census[census['Sex']==2][census['Year']==2000]['People']

#transforming to list
man_19=list(man_19)
women_19=list(women_19)
man_20=list(man_20)
women_20=list(women_20)

#population difference between genders from 1900
diff_1900=[]
for i in man_19:
  for j in women_19:
    if man_19.index(i) == women_19.index(j):
      diff_1900.append(i-j)
diff_1900

#population difference between genders from 2000
diff_2000=[]
for i in man_20:
  for j in women_20:
    if man_20.index(i) == women_20.index(j):
      diff_2000.append(i-j)
diff_2000

#importing pylab tools
from IPython.core.pylabtools import figsize
width=0.50
fig, ax = plt.subplots() # Creating figure and axis objects using subplots()

ax.bar(labels, man_19, width, label="Men_1900")
ax.bar(labels, women_19, width, bottom=man_19, label='Women_1900')
ax.set_xlabel("Age Groups")    #X-axis label
ax.set_ylabel("Population")    #Y-axis label
ax.set_title("Population by Gender from 1900")    #title
ax.legend()     #legend
plt.show()

from IPython.core.pylabtools import figsize
width=0.50
fig, ax = plt.subplots()    # Creating figure and axis objects using subplots()

ax.bar(labels, man_20, width, label="Men_2000")
ax.bar(labels, women_20, width, bottom=man_19, label='Women_2000')
ax.set_xlabel("Age Groups")   #X-axis label
ax.set_ylabel("Population")   #Y-axis label
ax.set_title("Population by Gender from 2000")    #title
ax.legend()   #legend
plt.show()
```
